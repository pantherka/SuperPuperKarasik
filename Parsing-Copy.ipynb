```python
from bs4 import BeautifulSoup
import re
import os
import sys
import gzip
from multiprocessing.dummy import Pool
import json
import codecs
from tqdm.notebook import tqdm
```

```python
def parse(file, index):
    info = {}
    info['id'] = index
    info['url'] = file.split('\n')[0]
    url = info['url']
    try:
        soup = BeautifulSoup(file.lower(), 'html5lib')
        title = soup.find('title')
        if title:
#             text = re.sub(r'[^\w\s]', '', title.text)
            info['title'] = re.sub(r"\s+", ' ', text)
        try:
            headings1 = soup.find_all('h1')
            if len(headings1) > 5:
                length = 5
            else:
                length = len(headings1)
            for i in range(length):
#                 text = re.sub(r'[^\w\s]','', headings1[i].text)
                info['h1_' + str(i)] = re.sub(r"\s+", ' ', text)
        except:
            message = "No h1 in " + url + " " + str(index)
            print(message, file=sys.stderr)
        try:
            headings2 = soup.find_all('h2')
            if len(headings2) > 5:
                length = 5
            else:
                length = len(headings2)
            for i in range(length):
#                 text = re.sub(r'[^\w\s]','', headings2[i].text)
                info['h2_' + str(i)] = re.sub(r"\s+", ' ', text)
        except:
            message = "No h2 in " + url + " " + str(index)
            print(message, file=sys.stderr)
        try:
            headings3 = soup.find_all('h3')
            if len(headings3) > 5:
                length = 5
            else:
                length = len(headings3)
            for i in range(length):
#                 text = re.sub(r'[^\w\s]','', headings3[i].text)
                info['h3_' + str(i)] = re.sub(r"\s+", ' ', text)
        except:
            message = "No h3 in " + url + " " + str(index)
            print(message, file=sys.stderr)
        try:
            p = soup.find_all('p')
            if len(p) > 10:
                length = 10
            else:
                length = len(p)
            for i in range(length):
#                 text = re.sub(r'[^\w\s]','', p[i].text)
                info['p_' + str(i)] = re.sub(r"\s+", ' ', text)
        except:
            message = "No p in: " + url + " " + str(index)
            print(message, file=sys.stderr)
        try:
            div = soup.find_all('div')
            if len(div) > 5:
                length = 5
            else:
                length = len(div)
            for i in range(length):
#                 text = re.sub(r'[^\w\s]','', div[i].text)
                info['div_' + str(i)] = re.sub(r"\s+", ' ', text)
        except:
            message = "No div in: " + url + " " + str(index)
            print(message, file=sys.stderr)
    except Exception as e:
        message = "Parsing error: " + url + " " + str(index)
        print(message, file=sys.stderr)
    return info
```
