```python
df_iter = pd.read_csv('./parsed_html.csv', encoding='utf-8', chunksize=500)
df = next(df_iter)
df = df.fillna('')
df = df.drop('title', axis=1)
df['div'] = df['div'].str.lower()
df['div'] = df['div'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
df['h1'] = df['h1'].str.lower()
df['h1'] = df['h1'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
df['h2'] = df['h2'].str.lower()
df['h2'] = df['h2'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
df['h3'] = df['h3'].str.lower()
df['h3'] = df['h3'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
df['p'] = df['p'].str.lower()
df['p'] = df['p'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
df['text'] = df[['div', 'h1', 'h2', 'h3', 'p']].apply(lambda x: ' '.join(x), axis = 1)
df = df.drop('div', axis=1)
df = df.drop('h1', axis=1)
df = df.drop('h2', axis=1)
df = df.drop('h3', axis=1)
df = df.drop('p', axis=1)
df['text'] = df['text'].map(lambda x: re.sub(' +',' ', x))
df.to_csv('project.csv', index=False)
```

```python
path = "project.csv"

for df in df_iter:
    df = df.fillna('')
    df = df.drop('title', axis=1)
    df['div'] = df['div'].str.lower()
    df['div'] = df['div'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
    df['h1'] = df['h1'].str.lower()
    df['h1'] = df['h1'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
    df['h2'] = df['h2'].str.lower()
    df['h2'] = df['h2'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
    df['h3'] = df['h3'].str.lower()
    df['h3'] = df['h3'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
    df['p'] = df['p'].str.lower()
    df['p'] = df['p'].map(lambda x: re.sub('[^а-яА-ЯёЁ]', ' ', x, count=0))
    df['text'] = df[['div', 'h1', 'h2', 'h3', 'p']].apply(lambda x: ' '.join(x), axis = 1)
    df = df.drop('div', axis=1)
    df = df.drop('h1', axis=1)
    df = df.drop('h2', axis=1)
    df = df.drop('h3', axis=1)
    df = df.drop('p', axis=1)
    df['text'] = df['text'].map(lambda x: re.sub(' +',' ', x))
    df.to_csv(path, mode='a', header=False, index=False)
```
